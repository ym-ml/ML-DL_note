```python
import numpy as np
import torch
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

----

```python
rand=np.array(torch.rand(100))
num=np.arange(len(rand))
data=pd.DataFrame({'x':num,'y':rand})
```

```python
data['y_nor']=(data['y']-data['y'].min())/(data['y'].max()-data['y'].min())
```

```python
data.head()
```

```python
fig,axes=plt.subplots(1,2,figsize=(10,4))
data.plot.scatter(x='x',y='y',ax=axes[0]),
data.plot.scatter(x='x',y='y_nor',ax=axes[1])
```

```python
#  ()
x = torch.normal(mean=3.0, std=0.5, size=(100,))

normal35=np.array(x)
num=np.arange(len(normal35))
data=pd.DataFrame({'x':num,'y':normal35})
```

```python
data['y_nor']=(data['y']-data['y'].min())/(data['y'].max()-data['y'].min())
```

```python
data.head()
```

```python
fig,axes=plt.subplots(1,2,figsize=(10,4))
data.plot.scatter(x='x',y='y',ax=axes[0])
data.plot.scatter(x='x',y='y_nor',ax=axes[1])
```

```python
#  [low, high)
x1 = torch.randint(low=0, high=5, size=(100,))
x2 = torch.normal(mean=3.0, std=0.5, size=(100,))

r=np.array(x2**x1)
num=np.arange(len(r))
data=pd.DataFrame({'x':num,'y':r})
```

```python
data['y_nor']=(data['y']-data['y'].min())/(data['y'].max()-data['y'].min())
```

```python
data.head()
```

```python
fig,axes=plt.subplots(1,2,figsize=(10,4))
data.plot.scatter(x='x',y='y',ax=axes[0])
data.plot.scatter(x='x',y='y_nor',ax=axes[1])
```
